```python
spend_reach = df.groupby(['Spend [USD]'])['Reach'].sum()
y_spend_reach = test.groupby(['Spend [USD]'])['Reach'].sum()
```

```python
plt.scatter(df['Spend [USD]'], df['Reach'], c='purple')
plt.scatter(test['Spend [USD]'], test['Reach'], c='salmon')
plt.legend(['Control', 'Test'])
plt.xlabel('Spend')
plt.ylabel('Reach')
plt.title('Comparison between Spend and Reach');
```

Undoubtedly, Control Campaign consists of the highest reach over the Test Campaign

```python
plt.scatter(df['# of Impressions'], df['Reach'], c='blue')
plt.scatter(test['# of Impressions'], test['Reach'], c='red')
plt.legend(['Control', 'Test'])
plt.xlabel('Impressions')
plt.ylabel('Reach')
plt.title('Comparison between Impressions and Reach');
```

There's a linear relationship between reach and impressions. However, the Control Campaign has a seemingly higher reach as well as impressions than the Test Campaign, insightful!

```python
plt.scatter(df['# of Website Clicks'], df['Reach'], c='gold')
plt.scatter(test['# of Website Clicks'], test['Reach'], c='green')
plt.legend(['Control', 'Test'])
plt.xlabel('Website Clicks')
plt.ylabel('Reach')
plt.title('Comparison between Website Clicks and Reach');
```

Surprisingly, the Test Campaign has got a relatively higher number of clicks than the Control Campaign. The Test Campaign has a lower level of reach, however, the Test Campaign has got a higher number of clicks than the Control Campaign

```python
plt.scatter(df['# of View Content'], df['Reach'], c='black')
plt.scatter(test['# of View Content'], test['Reach'], c='orange')
plt.legend(['Control', 'Test'])
plt.xlabel('View Content')
plt.ylabel('Reach')
plt.title('Comparison between View Content and Reach');
```

With a few outliers, Control Campaign seems to have more people that viewed the content, however, we cannot ascertain fully yet. We will attempt to acquire the sum of the people who viewed the content and estbalish which campaign was more effective in more getting people to view the content

```python
plt.scatter(sum(df['# of View Content']), sum(df['Reach']), c='black')
plt.scatter(sum(test['# of View Content']), sum(test['Reach']), c='orange')
plt.legend(['Control', 'Test'])
plt.xlabel('''Sum of 'View Content' ''')
plt.ylabel(''' Sum of 'Reach''')
plt.title('Comparison between View Content and Reach');
```

We can now establish with the above that Control Campaign was more effective in getting more people to view the content

```python
plt.scatter(df['# of View Content'], df['# of Add to Cart'], c='black')
plt.scatter(test['# of View Content'], test['# of Add to Cart'], c='orange')
plt.legend(['Control', 'Test'])
plt.xlabel('View Content')
plt.ylabel('Add to cart')
plt.title('Comparison between View Content and Add to cart');
```

```python
plt.scatter(df['Reach'], df['# of Add to Cart'], c='black')
plt.scatter(test['Reach'], test['# of Add to Cart'], c='orange')
plt.legend(['Control', 'Test'])
plt.xlabel('Reach')
plt.ylabel('Add to cart')
plt.title('Comparison between Reach and Add to cart');
```

More people from the Control Campaign added to cart than people from the Test Campaign

### Conclusion and Recommendation

It is safe to say that to establish either of the campaigns as the most effective, the goal would be the major determining factor.
If the goal is to create more brand awareness and perhaps to get some sales alongside, then the control campaign is most effective. However, if the goal is sales, the Test Campaign has proven to be slightly more effective. Although, the Test Campaign has a relatively lesser reach, hypothetically, if the reach could be increased, perhaps by upgrading the advertisement plan to reach more people or by including more advertisement channels, there is a potential that it could translate to a relatively higher conversion of advertisments into sales.

However, it is important to note that we cannot fully ascertain the amount of revenue that was generated from the purchases through this data, but, we can see in the data that the amount that was spent on both advertisements are largely higher than the number of purchases. This however does not directly translate to a profit or a loss, and to determine if or not the advertisment campaigns were profitable, further data on revenue will be required.
